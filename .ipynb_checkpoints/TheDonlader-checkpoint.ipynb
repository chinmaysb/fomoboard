```python
def initializeLists():
    pos_file = open("positive.txt", "r")
    neg_file = open("negative.txt", "r")
    pos_list = pos_file.readlines()
    pos_list = list(map(str.strip,pos_list))
    neg_list = neg_file.readlines()
    neg_list = list(map(str.strip,neg_list))
    pos_file.close()
    neg_file.close()
    return pos_list, neg_list

def posOrNeg(word):
    if(word in pos_list):
        return 1
    elif(word in neg_list):
        return -1
    else:
        return 0

def sentiment(sentence):
    tokenized_text = nltk.word_tokenize(sentence.lower())
    sentiment =0
    for word in list(set(tokenized_text)):
        sentiment += posOrNeg(word)
    return sentiment/len(tokenized_text)
```

```python
import csv
import re
from wordfreq import word_frequency
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string 

handle_scanner = re.compile(r'@[a-zA-Z0-9_]*')
hashtag_scanner = re.compile(r'#[a-zA-Z0-9_]*')
stop_words = set(stopwords.words('english'))

in_csv = open("theRealDonald.csv", "rt", encoding='utf-8', errors="ignore")
enriched_tweets = open('enriched_donald.csv', 'a', encoding='utf-16')

# create the csv writer object
csvwriter = csv.writer(enriched_tweets)
csvreader = csv.reader(in_csv)
count =0
lastLine = 13575
for row in csvreader:
    if(count<lastLine):
        pass
    elif(count==0):
        row.append('tagged handles')
        row.append('total followers of those tagged')
        row.append('hashtags')
        row.append('top keyword 1')
        row.append('top keyword 2')
        row.append('sentiment score')
        count+=1
        csvwriter.writerow(row)
    elif(count>0):
        handle_iterator = re.findall(handle_scanner, row[2])
        follower_count = 0
        tagged_handle = ""
        for handle in handle_iterator:
            follower_count += get_followers(handle)
            tagged_handle += handle + " | "
        row.append(tagged_handle)
        row.append(follower_count)
        hashtag_iterator = re.findall(hashtag_scanner, row[2])
        retweet_count = 0
        tagged_hashtag = ""
        for hashtag in hashtag_iterator:
            #retweet_count += get_retweets(hashtag)
            tagged_hashtag += hashtag + " | "
        row.append(tagged_hashtag)
        #row.append(retweet_count)        
        text_analyze = re.sub(r'[^\w\s]','',row[2])
        text_analyze = re.sub(r'http[\w/:.]*','',text_analyze)
        text_analyze = re.sub(r'@[a-zA-Z0-9]*','',text_analyze)
        text_analyze = re.sub(r'#[a-zA-Z0-9]*','',text_analyze)
        tokenized_text = nltk.word_tokenize(text_analyze.lower())
        tokenized_text = list(set(tokenized_text) - stop_words)
        freq_list = list()
        for word in tokenized_text:
                freq_list.append(word_frequency(word.lower(), 'en') * 1e6)
        try:
            row.append(tokenized_text[freq_list.index(min(freq_list))].lower())
            del tokenized_text[freq_list.index(min(freq_list))]
            del freq_list[freq_list.index(min(freq_list))]
            row.append(tokenized_text[freq_list.index(min(freq_list))].lower())
            row.append(sentiment(row[2]))
        except:
            pass
        csvwriter.writerow(row)
    count+=1
    if(count%150==0):
        print("Count as of now :" + str(count))
        #break
    
    #print(count)
        
in_csv.close()
enriched_tweets.close()
```
